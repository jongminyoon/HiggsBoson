```julia
function itp_ctop_kR(data::Array{Float64,3}, kR::Float64)
    ret = zeros(size(data)[1], size(data)[3])
    
    # interpolate ctop for the given (kR, af)
    for j=1:size(ret)[2]
        idx = findn(data[ctop,:,j])
        itp = interpolate(data[zR,idx,j], BSpline(Cubic(Line())), OnGrid())
        sitp = scale(itp, linspace(data[ctop,idx[1],j],data[ctop,idx[end],j],length(idx)))
        tempf(x) = sitp[x]
        ret[ctop,j] = nsolve(tempf, 1./kR ; lower = data[ctop,idx[1],j], upper = data[ctop,idx[end],j])
    end
    
    # interpolate other params for the given (ctop, af)
    for i=1:size(ret)[1]
        if i == ctop
            continue
        end
        for j=1:size(ret)[2]
            idx = findn(data[i,:,j])
            itp = interpolate(data[i,idx,j], BSpline(Cubic(Line())), OnGrid())
            sitp = scale(itp, linspace(data[ctop,idx[1],j],data[ctop,idx[end],j],length(idx)))
            ret[i,j] = sitp[ret[ctop,j]]
        end
    end
    return ret
end
;
```

```julia
itp_dat1 = zeros(3,8,af_len-1)
itp_dat2 = zeros(3,8,af_len-1)
for i=1:3
    itp_dat1[i,:,:] = itp_ctop_kR(data1[:,:,1:end-1], i+2.)
    itp_dat2[i,:,:] = itp_ctop_kR(data2[:,:,1:end-1], i+2.)
end
```

```julia
fig = plot(ylim = (0,20), xlim = (0.4, 0.7), xlabel = L"c_{top}", ylabel = L"a_f")
lcolor = [:red :green :blue :orange]
for i=1:3
    plot!(fig, itp_dat1[i,ctop,:], itp_dat1[i,af,:], label = "$(i+2) TeV", linecolor = lcolor[i])
    plot!(fig, itp_dat2[i,ctop,:], itp_dat2[i,af,:], label = "$(i+2) TeV", linestyle = :dash, linecolor = lcolor[i])
end
display(fig)
```
