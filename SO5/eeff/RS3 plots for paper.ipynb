```julia
@everywhere function eval_for_bottom(dev_func, f1, f2L, f2R, kR, v2f2, ct, cb)
    minphi = minphi_from_v2f2(v2f2)
    beta = find_beta(ct, cb)
    if string(f2L)[3] < string(f2R)[3] # e.g. bL4 and bR5
        beta = beta/sqrt(2)
    elseif string(f2L)[3] > string(f2R)[3] # e.g. bL5 and bR4
        beta = beta*sqrt(2)
    end
    
    if string(f2L)[3] == '4'
        ab = find_at4(ct, cb, beta, kR, minphi)
    elseif string(f2L)[3] == '5'
        ab = find_at5(ct, cb, beta, kR, minphi)
    else
        error("invalid f2")
    end
    
    arg2 = Dict("ct"=>ct, "cb"=>cb, "beta"=>beta, "ab"=>ab)
    return [evaluate(dev_func, f1, f2L, 0., arg2, p, kR, minphi) for p in plist], [evaluate(dev_func, f1, f2R, 0., arg2, p, kR, minphi) for p in plist]
end
```

```julia
@everywhere function eval_for_top(dev_func, f1, f2, kR, v2f2, ct, cb ; mixed = false, factor = sqrt(2))
    minphi = minphi_from_v2f2(v2f2)
    beta = find_beta(ct, cb)
    if mixed
        beta = beta * factor
    end
    
    if string(f2)[2] == '4'
        at = find_at4(ct, cb, beta, kR, minphi)
    elseif string(f2)[2] == '5'
        at = find_at5(ct, cb, beta, kR, minphi)
    else
        error("invalid f2")
    end
    arg2 = Dict("ct"=>ct, "cb"=>cb, "beta"=>beta, "at"=>at, "m"=>mt)
    result = [evaluate(dev_func, f1, f2, 0., arg2, p, kR, minphi) for p in plist]
    return [elem[1] for elem in result], [elem[2] for elem in result], [elem[3] for elem in result]
end
```
