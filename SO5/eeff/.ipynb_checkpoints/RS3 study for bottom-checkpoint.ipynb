### Generate data with the variables above

```julia
@everywhere function eval_for_bottom(dev_func, f1, f2L, f2R, kR, v2f2, ct, cb)
    minphi = minphi_from_v2f2(v2f2)
    beta = find_beta(ct, cb)
    if string(f2L)[3] < string(f2R)[3] # e.g. bL4 and bR5
        beta = beta/sqrt(2)
    elseif string(f2L)[3] > string(f2R)[3] # e.g. bL5 and bR4
        beta = beta*sqrt(2)
    end
    
    if string(f2L)[3] == '4'
        ab = find_at4(ct, cb, beta, kR, minphi)
    elseif string(f2L)[3] == '5'
        ab = find_at5(ct, cb, beta, kR, minphi)
    else
        error("invalid f2")
    end
    arg2 = Dict("ct"=>ct, "cb"=>cb, "beta"=>beta, "ab"=>ab)
    return [evaluate(dev_func, f1, f2L, 0., arg2, p, kR, minphi) for p in plist], [evaluate(dev_func, f1, f2R, 0., arg2, p, kR, minphi) for p in plist]
end
```

```julia
@time pmap(n -> begin
        for i=1:length(dev_func), j=1:length(f1), k=1:length(f2L), l=1:length(f2R)
            for m=1:length(kR), o=1:length(ct), p=1:length(cb)
                bottom[i,j,k,l,m,n,o,p,:,1], bottom[i,j,k,l,m,n,o,p,:,2] = 
                    eval_for_bottom(dev_func[i], f1[j], f2L[k], f2R[l], kR[m], v2f2[n], ct[o], cb[p])
            end
        end
        end, 1:length(v2f2))
;
```

```julia
# helper snippet to study from which deviations originate, between KK correction and dQ correction
decompose = quote
    args = Dict("ct"=>ct0, "cb"=>cb0, "beta"=>beta0, "ab"=>at0);
    function temp_decompose(dQ, KK)
        println(round.([dev_approx_eeff(f10, f20, 0., args, p, kR0, minphi0 ; dQ = dQ, KK = KK) for p in plist]; sigdigits = 4))
    end
    print("Full     : ")
    temp_decompose(1.,1.)
    print("KK only  : ")
    temp_decompose(0.,1.)
    print("dQ only  : ")
    temp_decompose(1.,0.)
    print("No KK&dQ : ")
    temp_decompose(0.,0.)
    println("   ")
end
;
```

### Limits on ct, cb from Ab and Rb

```julia
using LinearAlgebra

x0 = 0.
sx = 0.00519383
y0 = 0.
sy = 0.145727
corr = -0.926959

convmat = [sx^2 corr*sx*sy ; corr*sx*sy sy^2]
major = maximum(sqrt.(eigvals(convmat)))
minor = minimum(sqrt.(eigvals(convmat)))
theta = 0.5*atan(2*corr*sx*sy/(sx^2-sy^2))
p68 = sqrt(2.3)
p95 = sqrt(5.991)
t = collect(range(0, stop=2*pi, length=1001))
xaxis = sx > sy ? major : minor
yaxis = sx > sy ? minor : major

x68 = xaxis*p68*sin.(t)
y68 = yaxis*p68*cos.(t)
xx68 = x0.+([cos(theta) -sin(theta) ; sin(theta) cos(theta)]*[x68' ; y68'])[1,:]
yy68 = y0.+([cos(theta) -sin(theta) ; sin(theta) cos(theta)]*[x68' ; y68'])[2,:]

x95 = xaxis*p95*sin.(t)
y95 = yaxis*p95*cos.(t)
xx95 = x0.+([cos(theta) -sin(theta) ; sin(theta) cos(theta)]*[x95' ; y95'])[1,:]
yy95 = y0.+([cos(theta) -sin(theta) ; sin(theta) cos(theta)]*[x95' ; y95'])[2,:]
;
```

```julia
sw2 = gy^2/(g^2+gy^2)
cw2 = 1-sw2
function dg(c, a, T3L, T3R, Y, kR, v2f2)
    pre = normalization(z->f_zero(c,z),a ; norm_scope = 1)
    z2moment = pre*integrate_with_blkt(z->z^2*f_zero(c,z), a)
    logmoment = pre*integrate_with_blkt(z->2*z^2*log(1/z)*f_zero(c,z), a)
    return v2f2 * z2moment * (-T3L + T3R) / 2 / (cw2*T3L - sw2*Y) + mz^2/kR^2/4*(logmoment + z2moment)
end

dgL4(c, kR, v2f2) = dg(c, find_at0(c,kR,sin(minphi_from_v2f2(v2f2))^2), -1/2, 0, 1/6, kR, v2f2)
dgR4(c, kR, v2f2) = dg(-c, 0., 0., -1/2, -1/3, kR, v2f2)
dgL5(c, kR, v2f2) = dg(c, find_at0(c,kR,v2f2), -1/2, -1/2, 1/6, kR, v2f2)
dgR5(c, kR, v2f2) = dg(-c, 0., 0., 0., -1/3, kR, v2f2)
;
```
