```julia
@everywhere begin 

beta = 0.2
ce = 3.

d = Dict()
d["ct"] = [0.45, 0.5, 0.55, 0.6, 0.65]
d["kR"] = [2,3,4,5]
d["p"] = 0.2:0.2:1
d["tuning"] = [0.01, 0.05, 0.1]

end
;
```

```julia
@everywhere begin 

d["cb"] = findcb.(d["ct"],beta)

minphi_from_v2f2(tuning) = asin(sqrt(tuning))/2
d["minphi"] = minphi_from_v2f2.(d["tuning"])

d["af"] = [find_at5(ct, minphi, kR) for (ct, kR, minphi) in Iterators.product(d["ct"], d["kR"], d["minphi"])]

aL_from_kR_minphi(kR, minphi) = 0.5*sin(2*minphi)^2*(2/(1-zratio^2))/(mw^2/kR^2) - kpir
d["aL"] = [aL_from_kR_minphi(kR, minphi) for (kR, minphi) in Iterators.product(d["kR"], d["minphi"])]

iter = Iterators.product(1:length(d["kR"]), 1:length(d["minphi"]))

end
;
```

```julia
@everywhere begin 

# bL coupling
sme = [-1/2 -1.]
smq = [-1/2 -1/3]
rse = [-0.5 -0.5 0 0]
rsq = [-0.5 -0.5 0 2/3]

sm_bL = [sm(p, sme, smq) for p in d["p"]]
end



@everywhere function bL_loop(i)
    @printf "processing ct = %.2f \n" d["ct"][i]
    @time begin 
        ans = [100*(([geff(p, ce, d["ct"][i], d["aL"][j,k], d["aL"][j,k], d["kR"][j],rse,rsq ; af = d["af"][i,j,k]) for p in d["p"]]./sm_bL).^2 -1) for (j,k) in iter]
    end
    return ans
end

@time ebL5 = pmap(bL_loop, 1:length(d["ct"]))
;
```

```julia
@everywhere begin 

# bR coupling
sme = [-1/2 -1.]
smq = [0. -1/3]
rse = [-0.5 -0.5 0 0]
rsq = [0 0 0 -1/3]

sm_bR = [sm(p, sme, smq) for p in d["p"]]
end


@everywhere function bR_loop(i)
    @printf "processing ct = %.2f \n" d["ct"][i]
    @time begin 
        ans = [100*(([geff(p, ce, -d["cb"][i], d["aL"][j,k], d["aL"][j,k], d["kR"][j],rse,rsq ; af = 0. ) for p in d["p"]]./sm_bR ).^2 -1) for (j,k) in iter]
    end
    return ans
end

@time ebR5 = pmap(bR_loop, 1:length(d["ct"]))
;
```

```julia
function alphaw(aL)
    function totalintegral(x::Float64)
        p = -log(x)
        w = 1/(p^2*zratio)
        return p^3/x*3*(w/(G00(p)+aL*p*zratio*G10(p))/G11(p))^2
    end
    return quadgk(totalintegral,0,1;reltol=eps)[1]/(8pi^2)
end

function alphaz(aL,aB)
    sb2 = gy^2/g^2*(kpir+aB)/(kpir+aL);
    function totalintegral(x::Float64)
        p = -log(x)
        w = 1/(p^2*zratio)
        tG00 = G00(p); tG10 = G10(p); tG11 = G11(p);
        cup2 = (tG00 + aB*p*zratio*tG10) / ((tG00 + aB*p*zratio*tG10)+sb2*(tG00 + aL*p*zratio*tG10))
        return p^3/x*3/2*(w/(tG00 + aL*p*zratio*tG10)/tG11/cup2)^2
    end
    return quadgk(totalintegral,0,1;reltol=eps)[1]/(8pi^2)
end

function alphatop(c,af)
    function totalintegral(x::Float64)
        p = -log(x)
        w = 1/(p^2*zratio)
        return p^3/x*(-6)*(w/((Gmm(c,p) + af*p*zratio*Gpm(c,p))*(Gpp(c,p) + af*p*zratio*Gmp(c,p))/(1 + p^2*af^2*zratio^2)))^2
    end
    return quadgk(totalintegral,0,1;reltol=eps)[1]/(8pi^2)
end

function alphaT(c)
    function totalintegral(x::Float64)
        p = -log(x)
        w = 1/(p^2*zratio)
        return p^3/x*(-6)*(w/(Gpm(c,p)*Gmp(c,p)))^2
    end
    return quadgk(totalintegral,0,1;reltol=eps)[1]/(8pi^2)
end

function betaT(c)
    function totalintegral(x::Float64)
        p = -log(x)
        w = 1/(p^2*zratio)
        return p^3/x*(-6/4)*(w/(Gpm(c,p)*Gmp(c,p)))
    end
    return quadgk(totalintegral,0,1;reltol=eps)[1]/(8pi^2)
end


function estimatemh(ctop, cT, aL, aB, minphi, zR, af)
    return (-alphaw(aL) - alphaz(aL,aB) - alphatop(ctop,af) - alphaT(cT) + 2*betaT(cT))*8*0.5*sin(minphi)^2/zR^4/2/findff(minphi)
end
```

```julia
function solve1(ctop, aL, aB ; window = 0.03, maxphi = 0.5, upper = 1.5)
    
    lower = 0.;
    cT = (upper+lower)/2
    zR = 0.
    minphi = 0.
    higgsmass = 0.
    for tempcount=1:6     
        minphi = findminphi(ctop,cT,aL,aB ; level = 3)
        if minphi > maxphi
            upper = cT
        else
            zR = findzR(minphi,ctop)
            print("1          ")
            println(1/zR)
            higgsmass = findmh(ctop,cT,aL,aB,minphi,zR,findff(minphi))
            if higgsmass < mh
                upper = cT
            else
                lower = cT
            end
        end
        cT = (upper+lower)/2
    end

    for tempcount=1:3     
        minphi = findminphi(ctop,cT,aL,aB ; level = 4)
        if minphi > maxphi
            upper = cT
        else
            zR = findzR(minphi,ctop)
            print("2          ")
            println(1/zR)
            higgsmass = findmhinfty(ctop,cT,aL,aB,minphi,zR,findff(minphi))

            if higgsmass < mh
                upper = cT
            else
                lower = cT
            end
        end
        # println(cT, "           ", higgsmass)
        cT = (upper+lower)/2
    end

    minphi = findminphi(ctop,cT,aL,aB ; level = 5)
    zR = findzR(minphi,ctop)
            print("3          ")
            println(1/zR)
    higgsmass = findmhinfty(ctop,cT,aL,aB,minphi,zR,findff(minphi))

    for tempcount=1:15
        if zR > 0.00001
            soltemp = solution(2*log(1000/zR))

    global v = sqrt(soltemp[5]/2/soltemp[6])/1000
    global mw = sqrt(soltemp[2])/2*v
    global mz = sqrt(soltemp[2]+soltemp[1]*3/5)/2*v
    global mt = sqrt(soltemp[4]/2)*v
    global mh = sqrt(soltemp[5])/1000
    global g = 2*mw/v; global gy = sqrt((2*mz/v)^2-g^2)
        end

        if abs(higgsmass-mh) < window/1000
            break
        end

        if cT < 2e-4
            break
        end     
        
        if minphi > maxphi
            upper = cT
        else
            if higgsmass < mh
                upper = cT
            else
                lower = cT
            end
        end
        # println(cT, "           ", higgsmass)
        cT = (upper+lower)/2
        minphi = findminphi(ctop,cT,aL,aB ; level = 5)
        zR = findzR(minphi,ctop)
            print("4          ")
            println(1/zR)
        higgsmass = findmhinfty(ctop,cT,aL,aB,minphi,zR,findff(minphi))
    end
    

    return cT, minphi, zR, higgsmass
end

```
