## $ee \rightarrow tt$ study with varying $c_t$, and $v^2/f^2$

```julia
@everywhere begin
using SharedArrays
include("so5_eeff.jl");    

dev_func = [dev_approx_eett]
f1 = [eL5, eR5]
f2 = [t5approx, t5]
plist = [0.001, 0.35, 0.5, 0.75, 1.]
pscatter = [2,3,5]
kR = [1.5, 2.]
v2f2 = [0.01, 0.05, 0.1, 0.2]
ct = [0.2, 0.4, 0.6, 0.8]
cb = [0.2, 0.4, 0.6, 0.8] # needed to compute theta_b

end
top = SharedArray{Float64}(length(dev_func), length(f1), length(f2), 
    length(kR), length(v2f2), length(ct), length(cb), length(plist), 3)
;
```

```julia
mt = 0.173
mb = 0.00418
;
```

```julia
# check the approximate top quark wavefunctions are normalized correctly.
worst = 1.
for kR in [1.5,2.], minphi in [0.05, 0.1, 0.3]
#     println(kR, "  ", minphi)
    for ct in 0.3:0.1:0.7
#         println(ct)
        for cb in ct:0.1:1.0
            beta = find_beta(ct, cb)
            arg = Dict("ct"=>ct, "at"=>find_at5(ct, cb, beta, kR, minphi), "m"=>mt, "cb"=>cb, "beta"=>beta)
            q, func, at, qt = tL5approx(arg, kR, minphi)
            temp = sum(integrate_with_blkt(func,at)[1:2])
            if abs(temp-1) > abs(worst-1)
                worst = temp
            end
#             println(temp)
        end
    end
end
worst
```

### Generate data with the variables above

```julia
@everywhere function eval_for_top(dev_func, f1, f2, kR, v2f2, ct, cb ; mixed = false, factor = sqrt(2))
    minphi = minphi_from_v2f2(v2f2)
    beta = find_beta(ct, cb)
    if mixed
        beta = beta * factor
    end
    
    if string(f2)[2] == '4'
        at = find_at4(ct, cb, beta, kR, minphi)
    elseif string(f2)[2] == '5'
        at = find_at5(ct, cb, beta, kR, minphi)
    else
        error("invalid f2")
    end
    arg2 = Dict("ct"=>ct, "cb"=>cb, "beta"=>beta, "at"=>at, "m"=>mt)
    result = [evaluate(dev_func, f1, f2, 0., arg2, p, kR, minphi) for p in plist]
    return [elem[1] for elem in result], [elem[2] for elem in result], [elem[3] for elem in result]
end
```

```julia
@time pmap(n -> begin
        for i=1:length(dev_func), j=1:length(f1), k=1:length(f2)
            for m=1:length(kR), o=1:length(ct), p=1:length(cb)
                top[i,j,k,m,n,o,p,:,1], top[i,j,k,m,n,o,p,:,2], top[i,j,k,m,n,o,p,:,3] = 
                    eval_for_top(dev_func[i], f1[j], f2[k], kR[m], v2f2[n], ct[o], cb[p])
            end
        end
        end, 1:length(v2f2))
;
```

```julia
# helper snippet to study from which deviations originate, between KK correction and dQ correction
decompose = quote
    args = Dict("ct"=>ct0, "cb"=>cb0, "beta"=>beta0, "at"=>at0, "m"=>mt);
    function temp_decompose(dQ, KK)
        result = [dev_approx_eett(f10, f20, 0., args, p, kR0, minphi0 ; dQ = dQ, KK = KK) for p in plist]
        print(round.([elem[1] for elem in result][2:end]; sigdigits = 3))
        print(round.([elem[2] for elem in result][2:end]; sigdigits = 3))
        println(round.([elem[3] for elem in result][2:end]; sigdigits = 3))
    end
    println("Full     : ")
    temp_decompose(1.,1.)
    println("KK only  : ")
    temp_decompose(0.,1.)
    println("dQ only  : ")
    temp_decompose(1.,0.)
    println("No KK&dQ : ")
    temp_decompose(0.,0.)
    println("   ")
end
;
```
